# Step 0: Import Libraries

```python
import pandas as pd                                   #DataFrames
import numpy as np                                    #Numerical analysis
import matplotlib.pyplot as plt                       #Plotting
import seaborn as sns                                 #Visualization

from sklearn.preprocessing import PolynomialFeatures  #Polynomial Features for Linear Regressor
from sklearn.linear_model import LinearRegression     #Linear Regressor
```

# Step 1: Import Data

```python
#Import data
df = pd.read_csv("EconomiesOfScale.csv")
```

# Step 2: Visualize Data

```python
sns.jointplot(x='Number of Units', y='Manufacturing Cost', data = df, joint_kws={'color': '#F25C54'})
```

```python
sns.lmplot(x='Number of Units', y='Manufacturing Cost', \
           data=df, scatter_kws={'color': '#F25C54'}, line_kws={'color': '#392F5A'})
```

# Step 4: Train Model

```python
X=df[['Number of Units']]
y=df[['Manufacturing Cost']]

poly_regressor = PolynomialFeatures(degree=4)
X_columns = poly_regressor.fit_transform(X)
regressor = LinearRegression()
regressor.fit(X_columns, y)
```

# Step 5: Visualize Results

```python
y_predict = regressor.predict(poly_regressor.fit_transform(X))
```

```python
plt.scatter(X, y, color = '#F25C54')
plt.plot(X, y_predict, color = '#392F5A')
plt.ylabel('Cost Per Unit Sold (in USD)')
plt.xlabel('Number of Units (in Millions)')
plt.title('Unit Cost as a funciton of Number of Units')
```
